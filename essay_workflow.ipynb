```python
from langgraph.graph import StateGraph,START, END
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from typing import TypedDict
from pydantic import BaseModel, Field
from dotenv import load_dotenv
load_dotenv() 
```

```python
llm = HuggingFaceEndpoint(
    repo_id="google/gemma-2-2b-it",
    task="text-generation",
)

model = ChatHuggingFace(llm=llm)
```

```python
class EvaluationSchema(BaseModel):

    feedback: str = Field(description="Detailed feedback on the essay.")
    score: int = Field(description="score between 1 and 10 evaluating the quality of the essay.",ge=1,le=10)
```

```python
structured_model = model.with_structured_output(
    EvaluationSchema.model_json_schema(),
    method="json_mode"
)
```

```python
essay = """

The Rise of Artificial Intelligence and Its Impact on the Future

Artificial Intelligence (AI) has quickly become one of the most influential technologies in the world. From smartphones to self-driving cars, AI is shaping how we work, communicate, and solve problems. Although it is still developing, AI already plays a major role in our daily lives, and its importance will only grow in the future.

One of the biggest benefits of AI is its ability to process large amounts of information faster than humans. For example, AI can analyze thousands of medical images within seconds to help doctors detect diseases earlier. This makes healthcare faster, more accurate, and more accessible. In education, AI-powered tools can personalize learning for each student. These tools find a student’s strengths and weaknesses and then provide tailored lessons, helping students learn more effectively.

AI is also transforming industries by automating repetitive tasks. In factories, robots can perform dangerous or tiring work. In offices, AI helps with tasks like writing emails, analyzing data, or sorting information. This allows people to focus on creative and strategic work. However, some people worry that AI might replace jobs. While it is true that AI reduces certain types of work, it also creates new opportunities, especially in fields like robotics, data science, cybersecurity, and software development.

Despite the many advantages, AI also brings challenges. Privacy is a major concern, as AI systems often collect and analyze personal information. It is important for governments and companies to create strong rules that protect people’s data. Another challenge is ensuring that AI behaves fairly and does not discriminate. This requires careful programming and continuous monitoring.

In the future, AI will continue to grow, and its impact will be even greater. It could help solve global issues such as climate change by optimizing energy use, predicting natural disasters, and managing resources efficiently. It might also revolutionize transportation with safer self-driving vehicles and improve communication with real-time language translation.

In conclusion, Artificial Intelligence is a powerful tool that is transforming the world. It offers countless benefits in healthcare, education, and industry while also presenting challenges that must be addressed responsibly. As AI continues to evolve, it is important to use it wisely so that it helps humanity progress in a positive direction.

"""
```

```python
prompt = f"Evaluate the following essay and provide detailed feedback along with a score between 1 and 10 evaluating the quality of the essay:\n\n{essay}"

structured_model.invoke(prompt)
```
